```python
import re
import wikipedia

def scrape_description(query):
    print('scraping', query)
    text = wikipedia.summary(query, auto_suggest=False)
    text = re.sub(r'==.*?==+', '', text)
    text = text.replace('\n', '')
    return text
```

```python
wikipedia.search('beaver', suggestion=True)
```
